```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_ratings['ISBN'].value_counts()
```

```python
jumlah_rating = data_ratings.groupby('ISBN')['Book-Rating'].sum().reset_index()
```

```python
hasil = pd.merge(data_books, jumlah_rating, on='ISBN', how='left')
hasil.head()
```

```python
hasil_sort = hasil.sort_values(by = 'Book-Rating', ascending=False)
```

```python
hasil_sort = hasil_sort.head(40)
```

```python
hasil_sort
```

```python
hasil_sort.to_csv(index=False)
```

```python
sns.scatterplot(x='Book-Title', y='Book-Rating', data= hasil_sort)
plt.title('Top 20 Books')
plt.xlabel('Judul Buku')
plt.ylabel('Rating Buku')
plt.show()
```
